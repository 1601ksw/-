```python
#파일 불러오기

import csv
a=open('강남 서초.csv')
b=open('강동 송파.csv')
c=open('노원 도봉.csv')
d=open('동대문, 중랑.csv')
e=open('동작 관악.csv')
f=open('성동 광진.csv')
g=open('성북 강북.csv')
h=open('영등포 구로 금천.csv')
n=open('은평 서대문 마포.csv')
j=open('종로 중구 용산.csv')
k=open('강서 양천.csv')

#데이터 가공하기

result = []
data = []
price = []
student=[]
for file in [a,b,c,d,e,d,f,g,h,n,j,k] :
    tmp = csv.reader(file)
    for row in tmp :
        result.append(row)
result.sort()
for row in result :
    row[1] = row[1].replace(',','')
    row[2] = row[2].replace(',','')
    row[3] = row[3].replace(',','')
    if row[3] == '' or row[2] == '' or int(row[3]) * int(row[2]) == 0 or int(row[2])==1 or row[1]==0 :
        continue
    else :
        data.append(row[0].split(' ')[0])
        price.append((float(row[3]) / float(row[2]))*60)
        student.append(int(row[1]))
        
#학원비 평균
city = ['강남구', '강동구', '강북구' ,'강서구','관악구','광진구','구로구','금천구','노원구','도봉구','동대문구','동작구','마포구','서대문구','서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']        
city.sort()

cnt=0
list0=[]
for ad in city :
    average=0
    m=data.count(ad)
    for ae in range(cnt,cnt+m) :
        average=average+float(price[ae])
    list0.append(average/float(m))
    cnt=cnt+m


#학생수 합(수용 학생 수)

list1=[]
cmt=0
for ao in city :
    numst=0
    m=data.count(ao)
    for ap in range(cmt,cmt+m) :
        numst=numst+int(student[ap])
    list1.append(numst)
    cnt=cnt+m

#아파트 평당 가격 정렬

z=open('아파트 평당 가격.csv ')
Z=csv.reader(z)
tem=[]
list2=[]
for line in Z :
    tem.append(line)
tem.sort()
for rkw in tem :
    list2.append(rkw[1])

#파일 닫기

for file in [a,b,c,d,e,d,f,g,h,n,j,k,z] :
    file.close()
    
L0=[]
L1=[]
L2=[]
for qw in list0 :
    L0.append(float(qw)/float(list0[0]))
for qe in list1 :
    L1.append(float(qe)/float(list1[0]))
for qr in list2 :
    L2.append(float(qr)/float(list2[0]))
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family = 'Malgun Gothic')

plt.figure(dpi=300)
plt.plot(city,L0, label = '학원비')
plt.plot(city,L1, label = '인원')
plt.plot(city,L2, label = '매매가')
plt.title('구 별 수치 비교')
plt.legend(bbox_to_anchor=(1,0.8))
plt.xticks(city,rotation=70)
plt.show()
```
